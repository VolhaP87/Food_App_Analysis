```python
# For a high-level overview of the hotels, provide us the top 5 most voted hotels in the delivery category.

SELECT name, votes, rating FROM `zomato`
WHERE type='Delivery' 
ORDER BY votes DESC 
LIMIT 5;
```

```python
#The rating of a hotel is a key identifier in determining a restaurantâ€™s performance. Hence for a particular location called Banashankari find out the top 5 highly rated hotels in the delivery category.
SELECT name, rating, location, type FROM `zomato`
WHERE type='Delivery' AND location='Banashankari' 
ORDER BY rating DESC 
LIMIT 5;


Values:
    Onesta, 4.6, Banashankari, Delivery, 
    Ayodhya Upachar, 4.3, Banashankari, Delivery, 
    Corner House Ice Cream, 4.3, Banashankari, Delivery, 
    Mojo Pizza - 2X Toppings, 4.2, Banashankari, Delivery, 
    Smacznego, 4.2, Banashankari, Delivery, 
```

```python
# compare the ratings of the cheapest and most expensive hotels in Indiranagar.

SELECT rating AS rating1,
        (SELECT rating AS rating2
        FROM `zomato`
        WHERE location='Indiranagar'
        ORDER BY approx_cost DESC
        LIMIT 1)
    FROM `zomato` 
    WHERE location='Indiranagar' 
    ORDER BY approx_cost 
    LIMIT 1;


Headers: 
    rating1, rating2, 
    
    Values:
    3.7, 4.5, 
```

```python
# Online ordering of food has exponentially increased over time. 
# Compare the total votes of restaurants that provide online ordering services and 
# those who donâ€™t provide online ordering service.
SELECT SUM(votes) AS total_votes, online_order
FROM `zomato` 
GROUP BY online_order;

 Headers: 
    total_votes, online_order, 
    
    Values:
    1385573, Yes, 
    879446, No, 
```

```python
# Number of votes defines how much the customers are involved with the service provided by the restaurants
# For each Restaurant type, find out the number of restaurants, total votes, and average rating. 
# Display the data with the highest votes on the top( if the first row of output is NA display the remaining rows).
SELECT type, COUNT(type) AS number_of_restaurants, SUM(votes) AS total_votes, AVG(rating) AS avg_rating
FROM `zomato` 
WHERE type NOT LIKE 'NA'
GROUP BY type
ORDER BY total_votes DESC;

Values:
    Delivery, 5955, 697876, 3.27471, 
    Dine-out, 2162, 545242, 3.40870, 
    Buffet, 155, 60884, 3.65290, 
    Desserts, 485, 55675, 3.44619, 
    Cafes, 357, 50666, 3.33697, 
    Drinks & nightlife, 24, 16017, 3.56250, 
    Pubs and bars, 16, 9877, 3.93750, 
```
