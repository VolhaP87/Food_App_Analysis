```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('zomato.csv')
df
```

```python
df = df[['name', 'online_order', 'book_table', 'rate', 'votes', 'location', 'rest_type', 
         'dish_liked', 'cuisines', 'approx_cost(for two people)', 'listed_in(type)']]
df
```

```python
df = df.rename(columns = {'approx_cost(for two people)' : 'approx_cost',\
                          'listed_in(type)' : 'type',\
                          'rate' : 'rating'})
df
```

```python
df.info()
```

```python
df['online_order'] = df['online_order'].replace(np.nan, 'NA')
df['book_table'] = df['book_table'].replace(np.nan, 'NA')
df['rating'] = df['rating'].replace(np.nan, '0.0/5')
df['votes'] = df['votes'].replace(np.nan, '0')
df['location'] = df['location'].replace(np.nan, 'NA')
df['rest_type'] = df['rest_type'].replace(np.nan, 'NA')
df['dish_liked'] = df['dish_liked'].replace(np.nan, 'NA')
df['cuisines'] = df['cuisines'].replace(np.nan, 'NA')
df['approx_cost'] = df['approx_cost'].replace(np.nan, '0')
df['type'] = df['type'].replace(np.nan, 'NA')
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df = df[df["name"].str.contains("Rated|RATED") == False]
df = df[df["online_order"].str.contains("Rated|RATED") == False]
df = df[df["book_table"].str.contains("Rated|RATED") == False]
df = df[df["rating"].str.contains("Rated|RATED") == False]
df = df[df["votes"].str.contains("Rated|RATED") == False]
df = df[df["location"].str.contains("Rated|RATED") == False]
df = df[df["rest_type"].str.contains("Rated|RATED") == False]
df = df[df["dish_liked"].str.contains("Rated|RATED") == False]
df = df[df["cuisines"].str.contains("Rated|RATED") == False]
df = df[df["approx_cost"].str.contains("Rated|RATED") == False]
df = df[df["type"].str.contains("Rated|RATED") == False]
df
```

```python
df = df[df["rating"].str.contains("0.0/5") == False]
df
```

```python
df = df[(df['approx_cost']!='0')]
df
```

```python
df = df[(df['online_order']=='Yes') | (df['online_order']=='No')]
df
```

```python
df['rating'].unique()
```

```python
df = df[df["rating"].str.contains("-") == False]
df
```

```python
df['rating'] = df['rating'].replace('NEW', 0)
df
```

```python
df['rating'] = df['rating'].str.replace('/5', '')
df['rating'] = df['rating'].replace(np.nan, 0)
df
```
